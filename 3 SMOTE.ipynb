SMOTE (Synthetic Minority Over-sampling Technique) is a technique used in machine learning to address imbalanced datasets where the minority class has significantly fewer instances than the majority class. SMOTE involves generating synthetic instances of the minority class by interpolating between existing instances.

```python
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=1000,n_redundant=0,n_classes=2,n_features=2,n_clusters_per_class=2,weights=[0.90],random_state=12)
```

```python
import pandas as pd
df1 = pd.DataFrame(X,columns=['feature1','feature2'])
df2 = pd.DataFrame(y,columns=['target'])
df = pd.concat([df1,df2],axis=1)
```

```python
df.head()
```

```python
df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['feature1'],df['feature2'],c=df['target'])
```

```python
from imblearn.over_sampling import SMOTE
oversample = SMOTE()
X1,y1 = oversample.fit_resample(df[['feature1','feature2']],df['target'])
```

```python
df3 = pd.DataFrame(X1,columns=['feature1','feature2'])
df4 = pd.DataFrame(y1,columns=['target'])
df_oversampled = pd.concat([df3,df4],axis=1)
```

```python
plt.scatter(df_oversampled['feature1'],df_oversampled['feature2'],c=df_oversampled['target'])
```
