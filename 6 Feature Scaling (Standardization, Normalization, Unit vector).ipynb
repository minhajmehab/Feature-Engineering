## Feature Scaling

### 1 Standardization
### 2 Normalization (Min-Max Scaler)
### 3 Unit Vector

```python
import pandas as pd
import seaborn as sns
```

### 1 Standardization

```python
df = sns.load_dataset('tips')
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit_transform(df[['total_bill','tip']])
```

```python
pd.DataFrame(scaler.fit_transform(df[['total_bill','tip']]),columns=['total_bill_standardized','tip_standardized'])
```

### 2 Normalization (Min-Max Scaler)

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max = MinMaxScaler()
min_max.fit_transform(df[['total_bill','tip']])
```

```python
pd.DataFrame(min_max.fit_transform(df[['total_bill','tip']]),columns=['total_bill_normalized','tip_normalized'])
```

### 3 Unit Vector

```python
from sklearn.preprocessing import normalize
```

```python
pd.DataFrame(normalize(df[['total_bill','tip']]))
```
